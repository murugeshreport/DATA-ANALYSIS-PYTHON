```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../input/world-happiness-report-2021-worldwide-mortality/WHRData2021.csv')
```

```python
data
```

# Which Country More Population in 2019.? 

```python
mean = data['Population 2019'].mean()
data['Population 2019']=data['Population 2019'].fillna(mean)
```

```python
color = ['#191970','#8A2BE2','#8B008B','#9400D3','#C71585']
fig = plt.figure(figsize=(15,4))
sns.barplot(x=data.groupby(['Country name'])['Population 2019'].sum().index,y=data.groupby(['Country name'])['Population 2019'].sum().values,palette=color)
plt.tick_params(axis='x',rotation = 90,labelsize=7)
sns.despine()
plt.suptitle('Country Population Distripution in 2019',weight='bold',fontname='monospace',fontsize=20)
```

 Year **2019** 1st **INDIA** is more pupulation ,and 2nd **china** 

# Which Country More Population in 2020.?

```python
mean1 = data['Population 2020'].mean()
data['Population 2020']=data['Population 2020'].fillna(mean1)
```

```python
fig = plt.figure(figsize=(15,4))
sns.barplot(x=data.groupby(['Country name'])['Population 2020'].sum().index,y=data.groupby(['Country name'])['Population 2020'].sum().values,palette=color)
plt.tick_params(axis='x',rotation = 90,labelsize=7)
sns.despine()
plt.suptitle('Country Population Distripution in 2020',weight='bold',fontname='monospace',fontsize=20)
```

# How Much Differance Between Population 2019 and 2020?

```python
population2019 = data.groupby(['Country name'])['Population 2019'].sum().reset_index()
population2020 = data.groupby(['Country name'])['Population 2020'].sum().reset_index()
```

```python
totalpopulation=population2019.merge(population2020)
```

```python
totalpopulation['Differance_population']=totalpopulation['Population 2019'] - totalpopulation['Population 2020']
```

```python
x=totalpopulation['Country name']
y=totalpopulation['Population 2019']
x1=totalpopulation['Country name']
y1=totalpopulation['Population 2020']
```

```python
ax,legend=(['l1','l2'])
fig,ax =plt.subplots(figsize=(16,5))
ax.set_title('Difference Between population 2019 and  population 2020',weight='bold',fontsize=20)
ax.set_xlabel('Country',weight='semibold',rotation=90)
plt.tick_params(axis='x',rotation=90)
ax.plot(x,y)
ax.plot(x1,y1)
plt.legend(['population 2019','population 2020'])
```

# Which Country More of average distance ?

```python
mean2 = data['Log of average distance to SARS countries'].mean()
data['Log of average distance to SARS countries']=data['Log of average distance to SARS countries'].fillna(mean2)
```

```python
fig = plt.figure(figsize=(15,4))
sns.barplot(x=data.groupby(['Country name'])['Log of average distance to SARS countries'].sum().index,y=data.groupby(['Country name'])['Log of average distance to SARS countries'].sum().values,palette=color)
plt.tick_params(axis='x',rotation = 90,labelsize=7)
sns.despine()
plt.suptitle('Log of average distance to SARS countries Distripution',weight='bold',fontname='monospace',fontsize=20)
```

# which Country more Distance?top 10?

```python
top10_average_distance = data.groupby(['Country name'])['Log of average distance to SARS countries'].sum().reset_index().sort_values('Log of average distance to SARS countries',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(15,4))
sns.barplot(x=top10_average_distance['Country name'],y=top10_average_distance['Log of average distance to SARS countries'],palette=color)
plt.tick_params(axis='x',rotation = 90,labelsize=15)
sns.despine()
plt.suptitle('Top10_average_distance of Country',weight='bold',fontname='monospace',fontsize=20)
```

# Which country More Death in COVID-19 (2020)?

```python
mean4=data['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].mean()
data['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average']=data['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].fillna(mean4)
```

```python
fig = plt.figure(figsize=(15,6))
sns.barplot(x=data.groupby(['Country name'])['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].sum().index,y=data.groupby(['Country name'])['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].sum().values,palette=color)
plt.tick_params(axis='x',rotation = 90,labelsize=7)
sns.despine()
plt.suptitle('COVID -19 deaths in 2020 All country Distripution',weight='bold',fontname='monospace',fontsize=20)
```

# Which Country More death in COVID-19? and More affected?

```python
moredeath  =data.groupby(['Country name'])['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'].sum().reset_index().sort_values('Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(15,4))
sns.barplot(x=moredeath['Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average'],y=moredeath['Country name'],palette=color)
plt.tick_params(axis='x')
plt.xlabel('')
sns.despine()
plt.suptitle('More Affecated COV-19 Country',weight='bold',fontname='monospace',fontsize=20)
```
