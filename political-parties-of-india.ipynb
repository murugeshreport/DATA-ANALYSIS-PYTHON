# which year more political parties of india?

```python
data['Last_Year']=data['Last_Year'].fillna(0)
```

```python
data['Last_Year']=data['Last_Year'].astype(int)
```

```python
last = data['Last_Year'].value_counts().reset_index()
start = data['Start_Year'].value_counts().reset_index()
```

```python
last = last.astype(int)
start = start.astype(int)
```

```python
lastyr= last.drop('index',axis=1)
lastyr = lastyr.astype(int)
startyr = start.drop('index',axis=1)
startyr = startyr.astype(int)
```

```python
m = lastyr.join(startyr)
m['total'] = m['Start_Year'] + m['Last_Year']
```

```python
years = last.drop('Last_Year',axis=1)
years = years.astype(int)
years.columns=['years']
```

```python
yealperticipate = m.join(years)
```

```python
fig = plt.figure(figsize=(12,7))
sns.barplot(x=yealperticipate['years'],y=yealperticipate['total'],palette=color)
sns.despine()
plt.tick_params(axis='x',rotation = 90)
plt.title('All Year of Participate Political of India',weight = 'bold',fontsize=20)
```
