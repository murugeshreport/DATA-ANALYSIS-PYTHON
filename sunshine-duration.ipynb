```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../input/sunshine-duration-by-city/sunshine hours by city.csv')
```

```python
data.head()
```

# Year wise each country sunshine duration

```python
countrywise = data.groupby(['Country'])['Year'].sum()
```

# City Wise Yearly Sunshine Duration

```python
city  = data.groupby(['City'])['Year'].sum()
```

```python
plt.style.use('cyberpunk')
fig = plt.figure(figsize=(12,10))
plt.plot(city,color = 'red')
plt.tick_params(axis = 'x',rotation = 90,labelsize = 4)
plt.xlabel('City')
plt.title('Yearly City Sunchine Duration',)
mplcyberpunk.add_glow_effects()
```

# 1 QUATER Month Sunshine Duration of Each Country

```python
jan = data.groupby(['Country'])['Jan'].sum()
```

```python
feb = data.groupby(['Country'])['Feb'].sum()
Mar = data.groupby(['Country'])['Mar'].sum()
```

```python
plt.style.use('cyberpunk')
fig = plt.figure(figsize=(12,10))
plt.plot(jan,color = 'red')
plt.plot(feb,color = 'blue')
plt.plot(Mar,color = 'white')
plt.tick_params(axis = 'x',rotation = 90,labelsize = 4)
plt.xlabel('City')
plt.title('First Quater Month Sunchine Duration in Each Country',)
mplcyberpunk.add_glow_effects()
```
