```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
data =  pd.read_csv('../input/supermarket-sales/supermarket_sales - Sheet1.csv')
```

```python
data.head()
```

```python
df = data.groupby(['Product line'])[['Unit price','Quantity']].sum()
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection = '3d')
ax.scatter3D(data['Unit price'],data['cogs'],data['gross income'],color = 'red')
ax.set_xlabel('Unit Price')
ax.set_ylabel('Cogs')
ax.set_zlabel('gross income')
plt.show()
```

```python
g = data.groupby(['Quantity'])['Total'].sum().reset_index()
g
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111,projection = '3d')
x = g['Quantity']
y = g['Total']
z = np.zeros(10)
dx = np.ones(10)
dy = np.ones(10)
dz =np.ones(10)
ax.bar3d(x,y,z,dx,dy,dz,color = 'red')
plt.show()
ax.set_label('Quantity')
ax.set_label('Total')
```
