```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../input/sdfadgdadfda/medium_data.csv')
```

```python
data
```

```python
data.columns
```

```python
data[['year','month','day']]=data['date'].str.split('-',expand=True)
```

```python
color = ['#696969','#8B0000','#000000','#DC143C']
```

# High Claps Article

```python
data['claps'].max()
```

# Top Claps Articles

```python
claps = data['claps'].reset_index().sort_values('claps',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=claps['claps'],y=claps['claps'],palette=color)
plt.tick_params(axis='x',rotation=90)
plt.title('Top Claps article ',weight='bold',fontsize=20)
sns.despine()
```

# Top responses article

```python
responses = data['responses'].reset_index().sort_values('responses',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=responses['responses'],y=responses['responses'],palette=color)
plt.tick_params(axis='x',rotation=90)
plt.title('Top Responses article ',weight='bold',fontsize=20)
sns.despine()
```

# Top Reading time article

```python
reading_time = data['reading_time'].reset_index().sort_values('reading_time',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=reading_time['reading_time'],y=reading_time['reading_time'],palette=color)
plt.tick_params(axis='x',rotation=90)
plt.title('Top Reading time article ',weight='bold',fontsize=20)
sns.despine()
```

# Top title with claps

```python
title = data.groupby(['title'])['claps'].sum().reset_index().sort_values('claps',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=title['title'],y=title['claps'],palette=color)
plt.tick_params(axis='x',rotation=90)
plt.xlabel('Title',weight='bold',fontsize=15)
plt.title('Top 10 Title Claps',weight='bold',fontsize=20)
sns.despine()
```

# Top reading time with title

```python
 readingtime = data.groupby(['title'])['reading_time'].sum().reset_index().sort_values('reading_time',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=readingtime['title'],y=readingtime['reading_time'],palette=color)
plt.tick_params(axis='x',rotation=90)
plt.xlabel('Title',weight='bold',fontsize=15)
plt.title('Top 10 Title Readingtime',weight='bold',fontsize=20)
sns.despine()
```

# Top title with responses

```python
 responses = data.groupby(['title'])['responses'].sum().reset_index().sort_values('responses',ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=responses['title'],y=responses['responses'],palette=color)
plt.tick_params(axis='x',rotation=90)
plt.xlabel('Title',weight='bold',fontsize=15)
plt.title('Top 10 Title responses',weight='bold',fontsize=20)
sns.despine()
```

# Monthly publish article

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=data['month'].value_counts().index,y=data['month'].value_counts().values,palette=color)
plt.tick_params(axis='x',rotation=90)
plt.xlabel('Month',weight='bold',fontsize=15)
plt.title('Monthly Publish Article',weight='bold',fontsize=20)
sns.despine()
```

# Daily publish Article 

```python
fig = plt.figure(figsize=(12,6))
sns.barplot(x=data['day'].value_counts().reset_index().sort_values('index',ascending=False).index,
            y=data['day'].value_counts().values,palette=color)
plt.tick_params(axis='x',rotation=90)
plt.xlabel('day',weight='bold',fontsize=15)
plt.title('Daily Publish Articles',weight='bold',fontsize=20)
sns.despine()
```
