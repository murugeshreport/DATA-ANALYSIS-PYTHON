# Gender analysis

```python
gender=data['What is your gender? - Selected Choice'].value_counts().reset_index()
gender.columns=gender.columns=['gender','counts']
```

```python
percentage=gender['counts']/gender['counts'].sum()*100
percentage=pd.DataFrame(percentage).reset_index()
```

```python
gender=gender.join(percentage,lsuffix='_caller',rsuffix='percentage')
```

```python
gender.drop('index',axis=1,inplace=True)
```

```python
gender
```

```python
gender['counts_caller']=gender['counts_caller'].astype(int)
```

```python


gender.plot.bar(x='gender',y='counts_caller',color=color,figsize=(10,6),xlabel='Gender',ylabel='Counts of Gender',title=' Gender Distription')

```

```python
x=gender['counts_caller']
y=gender['gender']
fig=plt.figure(figsize=(10,8))
plt.pie(x,labels=y,colors=color,autopct='%.0f%%')
```
